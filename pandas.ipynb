```python
# ================================
# STEP 1: IMPORT LIBRARIES
# ================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ================================
# STEP 2: LOAD DATASET
# ================================
df = pd.read_csv("StudentPerformanceFactors.csv")

print("First 5 Rows:\n", df.head())

print("\nDataset Info:")
print(df.info())

# ================================
# STEP 3: BASIC DATA ANALYSIS
# ================================

# Select numeric columns
numeric_cols = df.select_dtypes(include='number')

# Calculate average of Exam Score
avg_score = df['Exam_Score'].mean()
print("\nAverage Exam Score:", avg_score)

# Summary statistics
print("\nStatistical Summary:\n", numeric_cols.describe())

# ================================
# STEP 4: BAR CHART
# ================================

# Example: Parental Involvement distribution
df['Parental_Involvement'].value_counts().plot(kind='bar')

plt.title("Parental Involvement Distribution")
plt.xlabel("Level")
plt.ylabel("Number of Students")
plt.show()

# ================================
# STEP 5: SCATTER PLOT
# ================================

# Hours Studied vs Exam Score
plt.scatter(df['Hours_Studied'], df['Exam_Score'])

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Hours Studied vs Exam Score")
plt.show()

# ================================
# STEP 6: HEATMAP (Correlation)
# ================================

corr = numeric_cols.corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

# ================================
# STEP 7: INSIGHTS
# ================================

print("\n--- INSIGHTS ---")

print("1. Average exam score indicates overall student performance level.")

print("2. Attendance shows strong positive correlation with exam scores (~0.58).")

print("3. Study hours also positively influence performance (~0.45 correlation).")

print("4. Previous scores and tutoring sessions have moderate impact.")

print("5. Sleep hours and physical activity show very weak correlation.")

print("6. Bar chart shows most students fall under 'Medium' parental involvement.")
```
